```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Download Tesla stock data
game_stop = yf.Ticker("GME")
game_stop_data = game_stop.history(period="1y")  # You can change the period as needed
```

```python
# Extract revenue data
revenue_data = game_stop.financials.loc['Total Revenue']
```

```python
# Plot revenue data
plt.figure(figsize=(10, 6))
revenue_data.plot(kind='bar')
plt.title('Game_Stop Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue (in billions)')
plt.show()
```
